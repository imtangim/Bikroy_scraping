```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset.isnull().sum()
```

```python
brandMode = dataset["Brand:"].mode()[0]
modelMode = dataset["Model:"].mode()[0]
yearMode = dataset["Year of Manufacture:"].mode()[0]
regeMode = dataset["Registration year:"].mode()[0]
conditionMode = dataset["Condition:"].mode()[0]
transMode = dataset["Transmission:"].mode()[0]
bodyMode = dataset["Body type:"].mode()[0]
fuelMode = dataset["Fuel type:"].mode()[0]
engineMode = dataset["Engine capacity:"].mode()[0]
milmode = dataset["Kilometers run:"].mode()[0]
pricemode = dataset["Price"].mode()[0]

dataset['Brand:']=dataset['Brand:'].fillna(brandMode)
dataset['Model:']=dataset['Model:'].fillna(modelMode)
dataset['Year of Manufacture:']=dataset['Year of Manufacture:'].fillna(yearMode)
dataset['Registration year:']=dataset['Registration year:'].fillna(regeMode)
dataset['Condition:']=dataset['Condition:'].fillna(conditionMode)
dataset['Transmission:']=dataset['Transmission:'].fillna(transMode)
dataset['Body type:']=dataset['Body type:'].fillna(bodyMode)
dataset['Fuel type:']=dataset['Fuel type:'].fillna(fuelMode)
dataset['Engine capacity:']=dataset['Engine capacity:'].fillna(engineMode)
dataset['Kilometers run:']=dataset['Kilometers run:'].fillna(milmode)
dataset['Price']=dataset['Price'].fillna(pricemode)


```
